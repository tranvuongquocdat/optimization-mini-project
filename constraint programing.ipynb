## Đề bài:

Có N bài báo 1, 2, …, N khoa học cần được phân cho M nhà khoa học
1, 2, …, M để nhận xét phản biện. Tùy thuộc vào lĩnh vực chuyên môn
của bài báo và của nhà khoa học mà mỗi bài báo i sẽ tương ứng với 1
danh sách L(i) các nhà khoa học có thể tham gia phản biện. Mỗi bài báo
cần được phân công cho ít nhất K nhà khoa học phản biện để đảm bảo
tính tin cậy. Hãy xây dựng phương án phân công phản biện sao cho số
lượng bài báo nhiều nhất phân cho một nhà khoa học nào đó phải ít
nhất.

• Input
• Dòng 1: ghi N, M, và K
• Dòng thứ i + 1 (i = 1,…, N): ghi giá trị qi và qi số nguyên dương trong
danh sách L(i)

```python
from ortools.linear_solver.pywraplp import Variable
from ortools.sat.python import cp_model
import itertools
import time
from generate_data import *

generate()
start_time = time.time()

N,M,K,A = read_data('data_sample.txt')
'''
print('N',N)
print('M',M)
print('K',K)
print('A',A)
'''

from ortools.sat.python import cp_model
class VarArraySolutionPrinter(cp_model.CpSolverSolutionCallback):
    def __init__(self,variables):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.__variables = variables
        self.__solution_count = 0
        
    def  on_solution_callback(self):
        self.__solution_count += 1
        for v in self.__variables:
            print('%s = %i'%(v,self.Value(v)),end = ' ')
        print()
    def solution_count():
        return self.__solution_count

model = cp_model.CpModel()

X = [[model.NewIntVar(0, 1, f'X[{i}][{j}]') for j in range(M)] for i in range(N)]
max_papers = model.NewIntVar(K,N+1,'max_papers')

#điểu kiện số chuyên gia nghiên cứu 1 bài báo lớn hơn hoặc bằng K
for i in range(N):
    model.Add(sum(X[i][j] for j in range(M)) >= K)

#điều kiện chuyên gia nghiên cứu bài báo phải thuộc đúng lĩnh vực đó
for i in range(N):
    for j in range(M):
        if A[i][j] == 0:
            model.Add(X[i][j] == 0)

#add constaint    
for j in range(M):
    model.Add(sum(X[i][j] for i in range(N)) <= max_papers)
    
model.Minimize(max_papers)
    
solver = cp_model.CpSolver()
status = solver.Solve(model)

if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    for j in range(M):
        print('chuyen gia' ,j+1,'thuc hien bai bao: ',end = '')
        for i in range(N):     
            if solver.BooleanValue(X[i][j]):
                print(i+1, end = ' ')
        print()
    print('max papers for each scientist = ', solver.Value(max_papers))
else:
    print('No solution found.')
'''
if status == cp_model.OPTIMAL or status == cp_model.FEASIBLE:
    for i in range(N):
        print('bai bao' ,i+1,'duoc thuc hien boi: ',end = '')
        for j in range(M):     
            if solver.BooleanValue(X[i][j]):
                print(j+1, end = ' ')
        print()
    print('max papers for each scientist = ', solver.Value(max_papers))
else:
    print('No solution found.')
'''

print('time:',round(time.time()-start_time,5),'s')
```
