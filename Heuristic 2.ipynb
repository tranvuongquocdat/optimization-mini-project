```python
import numpy as np
import time
#lấy dữ liệu từ data
#lấy dữ liệu từ data
def generate():
    import random

    N = int(input('so bai bao:'))
    M = int(input('so nha khoa hoc:'))
    K = int(input('so nha khoa hoc toi thieu nghien cuu 1 bai bao:'))

    a = [] #list chứa số nhà khoa học tham gia mỗi bài báo, K <= a[i] <= M
    for i in range(N):
        a.append(random.randint(K, M))   #so nha khoa hoc tham gia nghien cuu bai bao thu i

    X = []
    for x in a:
        X.append(random.sample(range(1,M+1), x))    #random những nhà khoa học nào sẽ thực hiện bài báo thứ i

    f = open('data_test.txt', 'w')

    f.write('{} {} {}\n'.format(N,M,K))
    for i in range(len(a)):
        f.write(str(a[i])+' ')
        f.writelines(str(X[i][j])+' ' for j in range (a[i]))
        f.write('\n')
    f.close()
generate()
start=time.time()
def outputfile(filename):
    with open(filename) as f:
        [N,M,K]=[int(x) for x in f.readline().split()]
        d=[]
        for i in range(N):
            d1=[int(x)-1 for x in f.readline().split()]
            d.append(list(d1[1:]))
        return N,M,K,d
N,M,K,d=outputfile('data_100.txt')
print(N,M,K)
#lấy dữ liệu về các nhà khoa học có thể tham gia những bài báo nào
def getdata(d):
    dict_data={i:[] for i in range(M)}
    for i in range(N):
        for j in d[i]:
             dict_data[j].append(i)
    len_data=[]
    for i in dict_data:
        len_data.append(len(dict_data[i]))
    len_d=[len(i) for i in d]
    data=[]
    for i in dict_data:
        data.append(dict_data[i])
    return data,len_data,d,np.array(len_d)
data,len_data,d,len_d=getdata(d)
#tạo ma trận M*N với M hàng là các M bài báo ,N cột  là  các nhà khoa học
# +new_X[i][j]=1 nếu bài báo i có thể được giá cho nhà bác học j(theo input) và =0 , ngược lại
new_X=[]
for i in range(N):
    row_X=[]
    for j in range(M):
        row_X.append(0) 
    new_X.append(row_X) 
for i in range(N):
    for j in d[i]:
        new_X[i][j]=1

def Heuristic():
    for i in range(N):
            for j in range(M):
                if new_X[i][j]==1:
                    max_col=max(len_data)
                    if (len_data[j]==max_col) and (len_d[i]>K):
                        new_X[i][j]=0
                        len_data[j]=len_data[j]-1
                        len_d[i]=len_d[i]-1
    final_solution={i:[] for i in range(M)}
    for i in range(N):
        for j in range(M):
            if new_X[i][j]==1:
                final_solution[j].append(i)
    optimal_solution=[]
    for i in final_solution: 
        print('nhà khoa học '+str(i) +' tham gia phản diện các bài báo:'+str(final_solution[i]))
        optimal_solution.append(len(final_solution[i]))
    print('optimal objective value is',max(optimal_solution))
Heuristic()
end=time.time()
print(end-start)
```
